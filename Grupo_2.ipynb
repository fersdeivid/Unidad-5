```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
from google.colab import files


uploaded = files.upload()  # Subir 'grupo1.csv'


df = pd.read_csv('grupo2.csv')


print("Primeras filas del archivo:")
print(df.head())

plt.figure(figsize=(8, 6))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df)
plt.title('Relación entre Horas de Estudio y Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.grid(True)
plt.show()


slope, intercept, r_value, p_value, std_err = linregress(df['HorasEstudio'], df['Calificacion'])


print(f"Ecuación de la recta: Calificación = {slope:.3f} * HorasEstudio + {intercept:.3f}")
print(f"Coeficiente de correlación (r): {r_value:.3f}")
print(f"Coeficiente de determinación (r²): {r_value**2:.3f}")
print(f"Valor p: {p_value:.3f}")


plt.figure(figsize=(8, 6))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df, label='Datos')
plt.plot(df['HorasEstudio'], intercept + slope * df['HorasEstudio'], color='red', label='Línea de regresión')
plt.title('Regresión Lineal: Horas de Estudio vs Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)
plt.show()
```
